```python
# ----------------------------------------------
# Missing Value Imputation Justification (Ratings)
# ----------------------------------------------

# For the 'rating' column, missing values were imputed using the **median**
# instead of the mean. This decision was based on the distribution of the 
# 'rating' data across different smartphone brands (e.g., OnePlus, Motorola, 
# Realme, Apple).

# Supporting reasons:
# - The skewness of the rating distributions for these brands is significantly
#   negative (left-skewed), with values such as:
#     - OnePlus: -1.18
#     - Motorola: -0.63
#     - Realme: -0.76
#     - Apple: -1.31
# - According to statistical conventions:
#     - Skewness between -0.5 and 0.5: ~Symmetric â†’ Mean is appropriate
#     - Skewness < -0.5 or > 0.5: Skewed â†’ Median is preferred
# - The KDE plots further confirm the skewness visually.
# - The **mean** can be heavily influenced by outliers in skewed distributions,
#   potentially leading to inaccurate imputations.
# - The **median**, being more robust to outliers and skewed data, provides a 
#   more accurate central value for imputation in such cases.

# Conclusion:
# â†’ Median imputation was chosen to maintain data integrity and avoid bias 
#   introduced by outliers or asymmetric distributions.
```

## ðŸ“Œ Conclusion

This exploratory data analysis (EDA) provided insightful observations about the smartphone market. Key findings from the analysis are:

1. **Missing Values**:  
   - Most missing values were handled appropriately.  
   - Ratings were imputed with the brand-wise median, while minor missing values like `num_cores` were filled manually.  
   - Some missing fields, like `processor_speed`, were not imputed due to ambiguity.

2. **Univariate and Bivariate Analysis**:  
   - The distribution of prices revealed that smartphones vary widely in cost, influenced by hardware specifications and brand value.
   - Features such as **5G support**, **NFC availability**, and **number of processor cores** showed strong correlation with higher price segments.
   - **Hexa-core** devices, mainly Appleâ€™s Bionic chips, were consistently more expensive than octa- or quad-core models.

3. **Brand and OS Impact**:  
   - Brands like **Apple**, **Samsung**, and **OnePlus** dominate the premium segment.
   - **iOS devices** had the highest average prices, followed by other operating systems, and then Android.
   - The brand and OS were strong indicators of device pricing and target market.

4. **Key Influencing Features on Price**:
   - **has_5g**, **has_nfc**, **processor_brand**, and **num_cores** were found to be the most influential features driving price.
   - Premium processors like **Apple Bionic**, **Kirin**, and **Google Tensor** are associated with higher prices.
   - **Fast charging**, though widely adopted, is not a major determinant of price.

---

### âœ… Overall Insights:

- ðŸ“ˆ **Price is Feature-Driven**: Smartphones are priced largely based on technical specifications. Performance-related attributes like 5G, processor type, and number of cores have a direct impact on pricing tiers.
  
- ðŸ·ï¸ **Brand Value Matters**: Apple devices, despite sometimes offering fewer raw specs, command premium prices due to strong brand value and ecosystem integration.

- ðŸ§  **Processor Brand as a Proxy for Market Segment**: Bionic and Kirin chipsets are almost exclusively found in high-end devices, while Snapdragon appears across all price segments, reflecting its versatility.

- ðŸ“² **Adoption of Modern Features**: Features like NFC, 5G, and high refresh rate displays are becoming standard in mid- to high-tier devices. Their presence serves as a reliable signal of device positioning.

- ðŸ“‰ **Budget Devices Show More Consistency**: Lower-priced phones tend to have simpler configurations with fewer performance features, leading to tighter price ranges and less variance.

- ðŸŒ **Market Segmentation is Clear**: The smartphone market can be cleanly divided into budget, mid-range, and premium segments based on just a few technical attributes and brand identifiers.
